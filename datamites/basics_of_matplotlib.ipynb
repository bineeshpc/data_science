```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
sales = [250, 300, 160, 90]
labels = ['Electronics', 'furnitures', 'office', 'home app']

plt.pie(sales, explode=[0, 0, 0.2, 0.2], labels=labels, autopct='%2.2f')
plt.savefig('storesales.png', format='png', transparent=True)

plt.show()
```

```python
from matplotlib import rcParams
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [7, 2, 6, 8, 1, 9]

rcParams['figure.figsize'] = 5, 4

```

```python
fig = plt.figure()
```

```python
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim([2, 5])

ax.set_xticks([2, 5])
ax.set_yticks([2, 5, np.mean(y)])

ax.plot(x, y)
```

```python
fig = plt.figure()
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y)
ax[1, 0].plot(x, y)
ax[0, 1].barh(x, y)
ax[1, 1].bar(x, y)
```

```python

sales = [20, 80, 60, 50]
year = [i for i in range(2016, 2020)]
profit = [0.5, 0.7, 0.5, 0.2]
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xticks(year)
plt.bar(year, sales, color=['red', 'blue'], width=profit)
plt.show()
plt.barh(year, sales)

plt.show()
```
