```python
import zipfile

```

```python
filename = "data/logs.zip"
```

```python
def zipfile_iterator(zf):
    fle = zipfile.ZipFile(zf)
    filenames = fle.namelist()
    for filename in filenames:
        print(filename)
        with fle.open(filename) as f:
            for line in f:
                yield line

for line in zipfile_iterator(filename):
    print(line)
```
