```python
# X input variables, SQFT

# X is capital by convention
# X is 2 dimensional

X = [ [1000], [1500], [1100]]

y = [60, 100, 90]

```

```python
# Scikit learn

from sklearn.linear_model import LinearRegression


model = LinearRegression()

model.fit(X, y)
```

```python
# m, slope as coefficient (a)

model.coef_
```

```python
print("Slope or coefficient a:", model.coef_)
print("Intercept, b:", model.intercept_)

model.predict([[2000]])
```

```python
model.predict([[2100]])
```

# Introducing multiple regression

ax + b = y

a -> coefficient 

b -> intercept

a1 x1 + a2 x2 + b = y

http://setosa.io/ev/ordinary-least-squares-regression/


```python
# Introducing multiple regression
# X -> Input variables, SQFT (x1), Floor (x2)
# the output depends on both variables

X = [[1000, 3], [1500, 4], [1100, 10]]
y = [60, 100, 90]

```

```python
# Scikit learn

from sklearn.linear_model import LinearRegression


model = LinearRegression()

model.fit(X, y)

model.predict([[1900, 3]])
```

```python
# Introducing multiple regression
# X -> Input variables, SQFT (x1), Floor (x2), number of bed rooms 


X = [[1000, 3, 5], [1500, 4, 4], [1100, 10, 3]]
y = [60, 100, 90]

```

```python
# Scikit learn

from sklearn.linear_model import LinearRegression


model = LinearRegression()

model.fit(X, y)

model.predict([[1900, 3, 3]])
```
